```
## Initialisation

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

#Range lat,long
# [81.90110,12.56700] to [41.87190,95.45946] mean 21.6+-6 78.1+-5 
latitude_list = df_events_data[df_events_data['state'] == 'AndhraPradesh']['latitude'].values
longitude_list = df_events_data[df_events_data['state'] == 'AndhraPradesh']['longitude'].values

#x0 = [[]]
#for i in range(len(latitude_list)):
#  llpair = "["+latitude_list[i]+","+longitude_list[i]+"]"
#  x0.append(llpair)

df = pd.DataFrame({
    'x': latitude_list,
    'y': longitude_list
})

df

np.random.seed(200)
k = 10
# centroids[i] = [x, y]
centroids = {
    i+1: [np.random.randint(12,21), np.random.randint(77, 85)]
    for i in range(k)
}

for i in centroids.items():
  print(i)

fig = plt.figure(figsize=(12, 12))
plt.scatter(df['x'], df['y'], color='grey')
colmap = {1: 'red', 2: 'green', 3: 'blue', 4: 'black', 5:'cyan', 6: 'yellow', 7: 'pink', 8: 'magenta', 9: 'orange', 10: 'brown'}
for i in centroids.keys():
    plt.scatter(*centroids[i], color=colmap[i])
plt.xlim(12, 21)
plt.ylim(77, 85)
plt.show()
```

```
def assignment(df, centroids):
    for i in centroids.keys():
        # sqrt((x1 - x2)^2 - (y1 - y2)^2)
        df['distance_from_{}'.format(i)] = (
            np.sqrt(
                (df['x'] - centroids[i][0]) ** 2
                + (df['y'] - centroids[i][1]) ** 2
            )
        )
    centroid_distance_cols = ['distance_from_{}'.format(i) for i in centroids.keys()]
    df['closest'] = df.loc[:, centroid_distance_cols].idxmin(axis=1)
    df['closest'] = df['closest'].map(lambda x: int(x.lstrip('distance_from_')))
    df['color'] = df['closest'].map(lambda x: colmap[x])
    return df

df = assignment(df, centroids)
print(df.head())

fig = plt.figure(figsize=(12, 12))
plt.scatter(df['x'], df['y'], color=df['color'], alpha=0.5, edgecolor='k')
for i in centroids.keys():
    plt.scatter(*centroids[i], color=colmap[i])
plt.xlim(12, 21)
plt.ylim(77, 85)
plt.show()
```

```
## Update Stage

import copy

old_centroids = copy.deepcopy(centroids)

def update(k):
    for i in centroids.keys():
        centroids[i][0] = np.mean(df[df['closest'] == i]['x'])
        centroids[i][1] = np.mean(df[df['closest'] == i]['y'])
    return k

centroids = update(centroids)
    
fig = plt.figure(figsize=(12, 12))
ax = plt.axes()
plt.scatter(df['x'], df['y'], color=df['color'], alpha=0.5, edgecolor='k')
for i in centroids.keys():
    plt.scatter(*centroids[i], color=colmap[i])
plt.xlim(12, 21)
plt.ylim(77, 85)
for i in old_centroids.keys():
    old_x = old_centroids[i][0]
    old_y = old_centroids[i][1]
    dx = (centroids[i][0] - old_centroids[i][0]) * 0.75
    dy = (centroids[i][1] - old_centroids[i][1]) * 0.75
    ax.arrow(old_x, old_y, dx, dy, head_width=2, head_length=3, fc=colmap[i], ec=colmap[i])
plt.show()
```

```
## Repeat Assigment Stage

df = assignment(df, centroids)

# Plot results
fig = plt.figure(figsize=(12, 12))
plt.scatter(df['x'], df['y'], color=df['color'], alpha=0.5, edgecolor='k')
for i in centroids.keys():
    plt.scatter(*centroids[i], color=colmap[i])
plt.xlim(12, 21)
plt.ylim(77, 85)
plt.show()
```

```
# Continue until all assigned categories don't change any more
while True:
    closest_centroids = df['closest'].copy(deep=True)
    centroids = update(centroids)
    df = assignment(df, centroids)
    if closest_centroids.equals(df['closest']):
        break

fig = plt.figure(figsize=(12, 12))
plt.scatter(df['x'], df['y'], color=df['color'], alpha=0.5, edgecolor='k')
for i in centroids.keys():
    plt.scatter(*centroids[i], color=colmap[i])
plt.xlim(12, 21)
plt.ylim(77, 85)
plt.show()
```
