```
# We fill missing values of latitude and longitude with average values
df_mean_lat_long = pd.DataFrame({'lat':0.0,'long':0.0},index=df_events_data[df_events_data['longitude'] == 0.0].groupby('city').count().index)


for x in df_events_data[df_events_data['longitude'] == 0.0].groupby('city').count().index:
  lat,long = df_events_data[df_events_data['city'] == x][['latitude','longitude']].mean()
  df_mean_lat_long.loc[x]['lat'] = lat
  df_mean_lat_long.loc[x]['long'] = long

#df_mean_lat_long[(df_mean_lat_long['lat'] > 13) & (df_mean_lat_long['lat'] < 30) ] 

filter_list = df_mean_lat_long.index

#df_events_data[df_events_data.city.isin(filter_list)].groupby('city').count()
  
#df_mean_lat_long 
for x in filter_list:
  print("City is:", x)
  df_events_data.loc[(df_events_data['city'] == x) & (df_events_data['latitude'] == 0.0),'latitude'] = df_mean_lat_long.loc[x]['lat']
  df_events_data.loc[(df_events_data['city'] == x) & (df_events_data['longitude'] == 0.0),'longitude'] = df_mean_lat_long.loc[x]['long']  
  
```
